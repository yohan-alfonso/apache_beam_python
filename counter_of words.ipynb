```python
import apache_beam as beam
import pandas as pd
import matplotlib.pyplot as plt
import apache_beam.dataframe as df
import apache_beam as beam

from apache_beam.io import ReadFromText, WriteToText
import apache_beam.runners.interactive.interactive_beam as ib
```

```python
import re as re

class WordExtractingDoFn(beam.DoFn):
      
      def process(self, element):
            return re.findall(r'[\w\']+', element, re.UNICODE) # encuentra todas las palabras de un texto
            
def run_pipeline(input_files):
      
      with  beam.Pipeline() as p1:
        
      #  lines = p1 | "ReadFiles" >> ReadFromText(input_files)
        

        word_count = (
          p1 | "ReadFiles" >> ReadFromText(input_files)
            
             | 'extract_words' >> beam.ParDo(WordExtractingDoFn())
             | 'PairWithOne' >> beam.Map(lambda x: (x, 1))
             | 'GroupAndSum' >> beam.CombinePerKey(sum)
             )

        word_count | beam.Map(print)

input_files = 'data.txt'
run_pipeline(input_files)
```
